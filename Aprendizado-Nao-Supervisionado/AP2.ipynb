```python
acidentes = pd.read_csv('./acidentes2022-top.csv')
acidentes.head()
```

```python
columns_drop = ['hora', 'ponto_controle', 'velocidade_max_via', 'divisao_via2', 'divisao_via3']
acidentes = acidentes.drop(columns_drop, axis = 1)
acidentes.head()
```

```python
contagem_nulos = acidentes.isnull().sum()
contagem_nulos
```

Algumas colunas foram excluídas do dataset (hora, ponto_controle, velocidade_max_via, divisao_via2, divisao_via3), por não conterem informações relevantes ou por ter um grande número de valores nulos.
Ainda foram mantidas algumas colunas com muitos valores nulos, por se entender que, devido à natureza da base de dados utilizada, alguns valores não necessitam ser preenchidos em determinadas ocorrências, mas, quando esses valores estão presentes, eles são relevantes para análise.

## Estrutura de grupos em 1d

```python
alt.Chart(acidentes).mark_point(
    size = 70
).encode(
    x = 'vitimas',
    tooltip = ['tempo_clima', 'vitimas']
).properties(
    width=600,
    height=150
).interactive()
```

```python
alt.Chart(acidentes).mark_point(
    size = 70
).encode(
    x = alt.X('vitimas'),
    tooltip = ['vitimas', 'tempo_clima', 'bairro']
).properties(
    width=600,
    height=150
).interactive()
```

## Estrutura de grupos em 2d



```python
alt.Chart(acidentes).mark_circle(
    size = 70
).encode(
    y = alt.Y('tipo', scale=alt.Scale(type='log')),
    x = 'acidente_verificado',
    tooltip = ['bairro', 'acidente_verificado', 'tipo']
).interactive()
```

```python
alt.Chart(acidentes).mark_circle(
    size = 70
).encode(
    y = alt.Y('tipo'),
    x = 'acidente_verificado',
    tooltip = ['bairro', 'acidente_verificado', 'tipo']
).interactive()
```

Obersava-se que, nesse contexto, não faz sentido utilizar a escala logarítmica, uam vez que ela é usada em gráficos quando os dados abrangem uma ampla faixa de valores, sendo especialmente útil quando há valores extremamente grandes ou quando se deseja destacar as diferenças em uma parte específica dos dados. Usa-se essa escala para lidar com grandes variações de valores, destacar pequenas diferenças, evitar distorções visuais, etc. Essas situações não se aplicam aos dados utilizados e, por isso, não há diferença na visualização aplicando tal escala.

```python
acidentes['Protocolo'] = acidentes['Protocolo'].astype(str)
acidentes['Acidente'] = acidentes['Protocolo'].str[4:].astype(int)

alt.Chart(acidentes).mark_point(
    size=70
).encode(
    x='Acidente',
    y='sinalizacao'
).interactive()
```

Imaginou-se uma relação entre as condições da sinalização e a quantidade de acidentes ocorridos no local. Para verificar, foi criado um gráfico que mostra o número de acidentes para cada situação indicada para a sinalização.

A manipulação inicial da coluna 'Protocolo' foi feita para excluir os digitos iniciais que indicavam o ano e manter apenas os dígitos finais, que especificavam cada acidente individualmente. Esse novo dado foi adicionado a uma coluna chamada Acidente.

Em seguida, avaliou-se a relação entre os dados mencionados. Observa-se, ao contrário do que se imagina, um número muito maior de acidentes com a sinalização em perfeito estado do que nos demais casos (Ilegível, Incompleta ou Não existente). Vale ressaltar, porém, que há também uma grande quantidade de casos em que a situação da sinalização não foi registrada (null), o que pode explicar esse resultado diferente do esperado.
