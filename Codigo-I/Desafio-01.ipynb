```python
# Ler arquivos

textoBase = '/content/drive/MyDrive/Programacao-Para-Aprendizagem-De-Maquina-BrunoDantas/MatrixRepresentation.txt'

arquivo = open(textoBase, 'r')
```

```python
# Uma matriz com todas as linhas

todasLinhas = []

with open(textoBase, "r") as arquivo:
  for i in arquivo:
    todasLinhas.append(i)  
```

```python
# [[],[],[]] = listaTexto[x]

listaLinhas = []

for l in todasLinhas:
  palavras = l.split()
  matriz = np.array(palavras)
  listaLinhas.append(matriz)
```

```python
#Matriz Resultado

matrizPrincipal = np.array(listaLinhas)

print(matrizPrincipal)
```
